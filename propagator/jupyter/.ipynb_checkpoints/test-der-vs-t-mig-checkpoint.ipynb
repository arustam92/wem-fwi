```python
%load_ext autoreload
%autoreload 2
import genericIO
import SepVector
import Hypercube
import numpy as np
import Operator
import time

import holoviews as hv
hv.extension('matplotlib')
```

```python
n1 = 100
n2 = 100 
n3 = 44

d1 = 0.02
d2 = 0.01
d3 = 1

o1 = 0
o2 = 0
o3 = 1

slowL = SepVector.getSepVector(Hypercube.hypercube(ns=[n1,n2,n3],ds=[d1,d2,d3],os=[o1,o2,o3]),storage='dataComplex')
slow2d = SepVector.getSepVector(Hypercube.hypercube(ns=[n1,n2],ds=[d1,d2],os=[o1,o2]),storage='dataComplex')

true = 2
bg = 1

slowNd = slowL.getNdArray()
slowNd[:].real = 1/true**2
# slowNd[:,int(1.5*n2/2)::,:].real = 1/3000*1/3000

slow2dNd = slow2d.getNdArray()
slow2dNd[:] = slowNd[0,:,:]
```

```python
import newExtWEM 
import json

par = {
    "ns" : 11,
    "dsx" : .2,
    "osx" : 0,
    "dsz" : 0.,
    "osz" : 0.5,
    
    "nr" : 100,
    "drx" : .02,
    "orx" : 0,
    "drz" : 0,
    "orz" : .9,
    "mode" : "dru",
    
    "fmax" : 15,
    "fmin" : 1,   
    
    "nref" : 1,
    "tap" : 100,
    "pad" : 100,
    "tpad" : 0,
    "ngs" : 3,
    "ngr" : 3,
    "prop" : 'ssf',
    
    "ntaylor" : 1,
    "eps" : .01
    
}


ot = -.5
tmax = 3
dt = 0.004
nt = int((tmax-ot) / dt)
wave = SepVector.getSepVector(Hypercube.hypercube(ns=[nt],ds=[dt],os=[ot]))
waveNd = wave.getNdArray()
t = np.linspace(ot,tmax,nt)
f0 = 5
waveNd[:] = (1-2*(np.pi*f0*t)**2)* np.exp(-(np.pi*f0*t)**2)
# waveNd[:] += np.flip(waveNd)
print(waveNd.shape)

data = SepVector.getSepVector(ns=[nt,par["nr"],par["ns"]],
                                  ds=[dt,par["drx"],par["dsx"]],
                                  os=[ot,par["orx"],par["osx"]])

parObj = genericIO.pythonParams(par)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure()
plt.plot(t,waveNd)

plt.figure()
plt.plot(slow2dNd[:,0])
```

```python
import WEM
wem = WEM.WEM(slow2d,data,wave,parObj)
wem.forward(False,slow2d,data)
```

```python
import numpy as np
f0 = par["fmin"]
df = 1/(tmax + 0.5 + par["tpad"]*dt)
im3d_time = np.zeros((nt+par["tpad"],im3d.shape[1],im3d.shape[2]),dtype=complex)
im3d_time[int(f0/df):n3+int(f0/df),:,:] = image3d[:]
im3d_time = np.fft.ifft(im3d_time,axis=0)
print(im3d_time.shape)
mat = np.fft.fftshift(im3d_time[:,:,50].T.real,axes=1)

plt.figure(figsize=[10,5])
plt.imshow(mat,cmap='gray',aspect=.001,extent=[-(tmax+0.5+ par["tpad"]*dt)/2,(tmax+.5+ par["tpad"]*dt)/2,0,1000])
plt.grid()
```

```python
x = np.linspace(-0.99,0.99,100)
xx = x**2
sqroot = 1/np.sqrt(1-xx)
# 	std::vector<float> coef = {1., 1./2. , 3./8. , 5./16. , 35./128.};

taylor1 = 1 + 1/2*xx
taylor2 = taylor1 + 3/8*xx**2
taylor3 = taylor2 + 5/16*xx**3
```

```python
plot = hv.Curve((x,sqroot)) * hv.Curve((x,taylor1)).opts(color='red') * hv.Curve((x,taylor2)).opts(color='cyan') * hv.Curve((x,taylor3)).opts(color='black')
plot.opts(ylim=(1,2))
```

## Derivative apply

```python
image3d_der = image3d.clone()
im3d_der = image3d_der.getNdArray()
tm = (tmax+.5+ par["tpad"]*dt)/2
der = Operator.Derivative(image3d,image3d,which=1,
                          order='exact',f0=par["fmin"],tmax=tm,dt=dt,alpha=0,beta=1)
der.forward(False,image3d,image3d_der)
der.adjoint(True,image3d_der, image3d)
```

```python
print(nt)
plt.plot(der.filter[:])
```

```python
fig, ax = plt.subplots(1,2,figsize=[20,10])
ax[0].imshow(im3d[:,:,50].T.real,aspect=.5)
ax[1].imshow(im3d_der[:,:,50].T.real,aspect=.5)
ax[0].grid(); ax[1].grid()
```

```python
fig, ax = plt.subplots(1,2,figsize=[10,2])
ax[0].plot(im3d[:,50,50].real,'red')
ax[0].plot(im3d_der[:,50,50].real,'green')
ax[0].grid()

# ax[1].plot(im3d[:,50,50].imag,'red')
ax[0].plot(im3d_der[:,50,50].imag,'green',linestyle='--')
ax[0].twinx().plot(np.angle(im3d_der[:,50,50]),'black')
ax[1].grid()
```

```python
import numpy as np
im3d_der_time = np.zeros((nt,im3d.shape[1],im3d.shape[2]),dtype=complex)
# positive frequencies
im3d_der_time[int(f0/df):n3+int(f0/df),:,:] = im3d_der[:]


im3d_der_time = np.fft.ifft(im3d_der_time,axis=0)
mat_der = np.fft.fftshift(im3d_der_time[:,:,50].T.real,axes=1)

fig, ax = plt.subplots(1,2,figsize=[20,10])
ax[0].imshow(mat,cmap='gray',aspect=.002,extent=[-tmax/2,tmax/2,0,1000])
ax[1].imshow(mat_der,cmap='gray',aspect=.002,extent=[-tmax/2,tmax/2,0,1000])
ax[0].grid()
ax[1].grid()
```

```python
tau = (t-1)
sc = 10
z0 = 50
plt.figure(figsize=[20,4])
plt.plot(t-1,sc*tau*mat[z0,:],'green',linestyle='--',label='t')
plt.plot(t-1,mat[z0,:],'blue',label='orig')
plt.plot(t-1,sc*mat_der[z0,:],'red',label='der')
plt.grid()
plt.legend()
```

## Second derivative

```python
image3d_der2 = image3d_der.clone()
der.adjoint(False,image3d_der2,image3d_der)
der.forward(True,image3d_der,image3d_der2)
```

```python
im3d_der2 = image3d_der2.getNdArray()
fig, ax = plt.subplots(1,2,figsize=[20,10])
ax[0].imshow(im3d[:,:,50].T.real,aspect=.5)
ax[1].imshow(im3d_der2[:,:,50].T.real,aspect=.5)
ax[0].grid(); ax[1].grid()
```

```python
fig, ax = plt.subplots(1,2,figsize=[10,2])
ax[0].plot(im3d[:,50,50].real,'red')
ax[0].plot(im3d_der2[:,50,50].real,'green')
ax[0].grid()

ax[1].plot(im3d[:,50,50].imag,'red')
ax[0].plot(im3d_der2[:,50,50].imag,'green',linestyle='--')
ax[1].grid()
```

```python
import numpy as np
im3d_der2_time = np.zeros((nt,im3d.shape[1],im3d.shape[2]),dtype=complex)
im3d_der2_time[int(f0/df):n3+int(f0/df),:,:] = im3d_der2[:]
im3d_der2_time = np.fft.ifft(im3d_der2_time,axis=0)
mat_der2 = np.fft.fftshift(im3d_der2_time[:,:,50].T.real,axes=1)

fig, ax = plt.subplots(1,2,figsize=[20,10])
ax[0].imshow(mat,cmap='gray',aspect=.002,extent=[-tmax/2,tmax/2,0,1000])
ax[1].imshow(mat_der2,cmap='gray',aspect=.002,extent=[-tmax/2,tmax/2,0,1000])
ax[0].grid()
ax[1].grid()
```

```python
tau = (t-1)
sc = 10
z0 = 50
plt.figure(figsize=[20,4])
plt.plot(t-1,sc*tau*tau*mat[z0,:],'green',linestyle='--',label='t2')
plt.plot(t-1,mat[z0,:],'blue',label='orig')
plt.plot(t-1,sc*mat_der2[z0,:],'black',label='der2')
plt.grid()
plt.legend()
```
